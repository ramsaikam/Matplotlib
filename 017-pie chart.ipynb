```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use("default")
%matplotlib inline
```

```python
City = ["Hyd","Bangalore","Chennai","Delhi","Kolkata","Mumbai","Ahmedabad"]
Millionaire = [500,1200,600,2000,400,3000,250]
colors = ["red","green","blue","grey","yellow","chocolate","orange"]
```

```python
fig,ax = plt.subplots()
fig.set_figwidth(10)
fig.set_figheight(6)


explodes = np.zeros(7)
explodes[1] = 0.1
explodes[3] = 0.008          # Centre is at (0, 0), Default pie radius = 1
explodes[5] = 0.07


def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct



ax.pie(Millionaire,            # slice  data
        labels = City,         # slice labels
        colors = colors,       # slice colors
        startangle=90,         # starting angle for first one in acw direction.
        shadow = True,         # Enable or disable shadow
        explode = explodes,    # Explode pieces
        radius = 1,            # Radius of pie chart          
        autopct = make_autopct(Millionaire), 
       #autopct = "%2.2f%%",   # percentage format to display
        pctdistance=0.8,       # distance from centre to display percentage 
        labeldistance=1.1,     # distance from centre to display label
        counterclock=False,    # pie chart formation sense
        frame=True             # fancy frame
      )         

ax.axis([-1.5,1.5,-1.5,1.5])
ax.set_title("Millionaires in each city")
#ax.axis('equal')                   # to make the chart circular
plt.show()
```
