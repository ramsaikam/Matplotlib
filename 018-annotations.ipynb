```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use("default")
```

```python
days = np.arange(1,8)
sales_hyd = [20,40,60,80,90,35,10]
sales_bangalore = [40,55,68,100,120,65,40]
```

```python
fig,ax = plt.subplots()
fig.set_figwidth(10)
fig.set_figheight(8)


ax.plot(days,sales_hyd,color = "green",label = "Hyderabad")
ax.plot(days,sales_bangalore,color = "red",label = "Bangalore")




ax.plot(days[sales_hyd.index(max(sales_hyd))],
        max(sales_hyd),
        color = "blue",
        marker = "*",markersize = 20)

ax.plot(days[sales_bangalore.index(max(sales_bangalore))],
        max(sales_bangalore),
        color = "black",
        marker = ".",markersize = 20)




ax.annotate(s = "maximum-sales-hyd",   #text to display
             xy = (days[sales_hyd.index(max(sales_hyd))],
                           max(sales_hyd)),  #point to annote
             xytext =(days[sales_hyd.index(max(sales_hyd))]-0.5,
                           max(sales_hyd)+5 ),# the coordinate for the text to start
            arrowprops=dict(facecolor='grey', shrink=1.0))
            
ax.annotate(s = "maximum-sales-bangaloe",   #text to display
             xy = (days[sales_bangalore.index(max(sales_bangalore))],
                           max(sales_bangalore)),      #point to annote
             xytext =(days[sales_bangalore.index(max(sales_bangalore))]-0.5,
                           max(sales_bangalore)+2 ),# the coordinate for the text to start)
           arrowprops=dict(facecolor='grey', shrink=1.0))




ax.set_xlabel("Days")
ax.set_ylabel("Sales of TV's")
ax.set_title("Sales comparision in each city")
ax.legend(loc = "upper left")
ax.grid()
plt.show()
```
