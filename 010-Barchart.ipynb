```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use("default")
```

```python
range_vals = np.linspace(0, 100, 11)
counts1 = np.random.rand(10)*4.0
errors = np.ones(10)*0.5
bar_Width = 3.0

mid_vals = (range_vals[0:-1]+range_vals[1:])*0.5 - bar_Width*0.5# for classic style
#mid_vals = (range_vals[0:-1]+range_vals[1:])*0.5 # for default style
```

```python
fig,ax = plt.subplots(figsize = (12,10))

ax.bar(mid_vals,counts1,color='chocolate', width=bar_Width, 
                                   label='sample bar1',yerr = errors)



ax.set_xticks(mid_vals+bar_Width*0.5) # for classic style
#ax.set_xticks(mid_vals) # for default style

ax.set_xlabel("Values")
ax.set_ylabel("Count/fraction")
ax.set_title("Bar chart")
ax.grid()
ax.legend()
plt.show()
```
