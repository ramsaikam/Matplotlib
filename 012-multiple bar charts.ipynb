```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use("default")
%matplotlib inline
```

```python
label = ['Adventure', 'Action', 'Drama', 'Comedy', 'Horror', 'Musical']
India = [941,854,4595,2125,942,509]
USA = [1000,1500,500,600,4000,178]
Chaina = [2000,5000,500,150,234,400]
numbers = np.arange(6)
bar_width = 0.3
```

```python
fig,ax = plt.subplots(figsize = (15,8))
india = ax.bar(numbers,India,label = "India",color = "red",width = bar_width)
usa = ax.bar(numbers+bar_width,USA,label = "USA",color = "blue",width = bar_width)
chaina = ax.bar(numbers+2*(bar_width),Chaina,label = "Chaina",color = "green",width=bar_width)
ax.set_xticklabels(label,rotation = 90)
ax.set_xticks((numbers+bar_width))



ax.set_xlabel("Movie type")
ax.set_ylabel("No of movies")
ax.set_title("Movies in a year")
#ax.grid()
ax.legend(loc = "best")



def autolabel(country):
    
    """
    Attach a text label above each bar displaying its height
    """
    for country_type in country:
        height = country_type.get_height()
        ax.text((country_type.get_x() + country_type.get_width()) - 0.2 ,
                               height * 1.025,
                              '%d' % int(height),
                                ha='center', 
                                va='bottom',
                                fontweight = "bold")
        
autolabel(india)
autolabel(usa)
autolabel(chaina)


plt.show()
```
