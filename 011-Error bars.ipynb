```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use("default")
%matplotlib inline
```

```python
label = ['Adventure', 'Action', 'Drama', 'Comedy', 'Horror', 'Musical']
India = [941,854,4595,2125,942,509]
variance = [120,100,500,250,100,70]
numbers = np.arange(6)
bar_width = 0.4
```

```python
fig,ax = plt.subplots(figsize = (10,8))
ax.bar(numbers,India,label = "India",color = "red",width=bar_width,yerr = variance)
ax.set_xticklabels(label,rotation = 90)
ax.set_xticks(numbers)


ax.set_xlabel("Movie type")
ax.set_ylabel("No of movies")
ax.set_title("Movies in a year")
ax.grid()
ax.legend(loc = "best")
plt.show()
```
