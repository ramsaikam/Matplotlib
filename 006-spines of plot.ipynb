```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-6,6,1000)
y = np.sin(x)
z = np.cos(x)
```

```python
fig,ax = plt.subplots()
ax.plot(x,y,color = "red",label = "sin")
ax.plot(x,z,color = "green",label = "cos")

#put x-axis up
#plt.axes().spines['bottom'].set_position(('data', 0))
ax.spines["bottom"].set_position(("data",0))

# push the y axis right
#plt.axes().spines['left'].set_position(('data', 0))
ax.spines["left"].set_position(("data",0))



# add a horizontal line at the bottom
# plt.axhline(y=-1)
#plt.axhline(y=plt.ylim()[0], color='k') # plt.ylim() returns a tuple (ymin, ymax)
ax.axhline(y= plt.ylim()[0],color = "red")


# add a vertical line at the left
# plt.axvline(x=-7)
#plt.axvline(x=plt.xlim()[0], color='black') # plt.xlim() returns a tuple (xmin, xmax)
ax.axvline(x= plt.xlim()[0],color = "green")

ax.legend()
ax.grid()
ax.set_xlabel("angle")
ax.set_ylabel("magnitude")
ax.set_title("plot of sine and cosine")
plt.show()
```
