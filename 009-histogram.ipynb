```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use("ggplot")
%matplotlib inline
```

```python
mean = 100
sd = 15
samples = 10000
bin_size = 15
IQ_score= np.random.normal(mean,sd,samples)
```

```python
fig,ax = plt.subplots(figsize = (10,5))
#ax.hist(IQ_score,bins=bin_size,facecolor = "red",label = "IQ's",normed=True)
counts,bins,extras = ax.hist(IQ_score,
                             bins=bin_size,
                             facecolor = "red",
                             label = "IQ's",
                             normed=False) #True

ax.plot((bins[1:]+bins[0:-1])*0.5, counts, 
         label='series', 
         color='xkcd:navy blue')

ax.set_xticks(bins)
ax.set_xlabel("IQ score")
ax.set_ylabel("Count/Fraction")
ax.set_title("Indian IQ score")
ax.legend()
ax.grid()
plt.show()
```
