```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.array([10,12,15,16,18,20])
y = np.array([100,120,150,160,180,200])

x_numbers = np.array([8,10,12,14,16,18,20,22])
y_numbers = np.array([90,110,130,150,170,190,210])
```

```python
plt.plot(x,y,label = "Sales in hyd")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales vs Days")
plt.xticks(x_numbers)
plt.yticks(y_numbers)
plt.legend()
plt.grid()
plt.axis([8,22,90,220])
plt.show()
```

```python
#2nd type

fig,ax = plt.subplots()

ax.plot(x,y,label = "Sales in HYD")
ax.set_xlabel("Days")
ax.set_ylabel("Sales")
ax.set_title("Sales vs Days")
ax.set_xticks(x_numbers)
ax.set_yticks(y_numbers)
ax.axis([8,21,90,220])
ax.legend()
ax.grid()
plt.show()
```
