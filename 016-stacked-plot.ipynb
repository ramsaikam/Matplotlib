```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use('default')
```

```python
np.random.seed(111)
years = np.linspace(2001,2010,10)
expenditure_food = np.random.randint(10,40,10)
expenditure_clothes = np.random.randint(5,10,10)
expenditure_health = np.random.randint(10,20,10)
expenditure_tours = np.random.randint(15,20,10)
```

```python
expenditure = [expenditure_food,expenditure_clothes,expenditure_health,expenditure_tours]
```

```python
labels = ["food","clothes","health","tours"]
colors = ["red","green","blue","yellow"]
```

```python
fig,ax = plt.subplots()
fig.set_figwidth(10)
fig.set_figheight(7)


ax.stackplot(years,
             expenditure,
            colors = colors,
            labels = labels)
ax.legend(loc = "best")
ax.set_xlabel("Years")
ax.set_ylabel("accumulated expenditue in(lakhs)")
ax.set_title("stacked plot:The expenditure in {} city for 10 years in {} {}".
                        format("Bangalore","lakh","rupees"))
plt.show()
```
