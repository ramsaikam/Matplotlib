```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use("default")
```

```python
x = np.linspace(-2.0*np.pi, 2.0*np.pi, 100)
sin_values = np.sin(x) 
cos_values = np.cos(x)
```

```python
fig,ax = plt.subplots(4,2)
fig.set_figwidth(10)
fig.set_figheight(20)

#upper value fill
# Simply: area above the curve from a reference above the curve
ax[0][0].plot(x,sin_values,color = "red" ,label = "sin")
ax[0][0].plot(x,cos_values,color = "green" ,label = "cos")
ax[0][0].fill_between(x,sin_values,1)
ax[0][0].fill_between(x,cos_values,1)
ax[0][0].set(title = "upper value fill",
       xlabel = "x",
       ylabel = "sine values")
ax[0][0].legend(loc = "upper left")



#mid value fill
# Simply: area under the curve with the reference intersecting the curve
ax[0][1].plot(x,sin_values,color = "red" ,label = "sin")
ax[0][1].plot(x,cos_values,color = "green" ,label = "cos")
ax[0][1].fill_between(x,0,sin_values)
ax[0][1].fill_between(x,0,cos_values)

ax[0][1].set(title = "mid value fill",
       xlabel = "x",
       ylabel = "sine values")
ax[0][1].legend(loc = "upper left")



#lower value fill
# Simply: area below the curve from a reference below the curve
ax[1][0].plot(x,sin_values,color = "red" ,label = "sin")
ax[1][0].plot(x,cos_values,color = "green" ,label = "cos")

ax[1][0].fill_between(x,-1,sin_values)
ax[1][0].fill_between(x,-1,cos_values)
ax[1][0].set(title = "lower value fill",
       xlabel = "x",
       ylabel = "sine values")
ax[1][0].legend(loc = "upper left")


#area between curves
# Simply: Area between the curves
ax[1][1].plot(x,sin_values,color = "red" ,label = "sin")
ax[1][1].plot(x,cos_values,color = "green" ,label = "cos")

ax[1][1].fill_between(x,sin_values,cos_values,color = "grey")

ax[1][1].set(title = "area between curves",
       xlabel = "x",
       ylabel = "sine values")
ax[1][1].legend(loc = "upper left")



# Area between curves
# Limited regions

ax[2][0].plot(x,sin_values,color = "red" ,label = "sin")
ax[2][0].plot(x,cos_values,color = "green" ,label = "cos")

ax[2][0].fill_between(x[20:60],sin_values[20:60],cos_values[20:60],
                        color = "yellow")

ax[2][0].set(title = "area between curves but limited",
       xlabel = "x",
       ylabel = "sine values")
ax[2][0].legend(loc = "upper left")


# Using where option
ax[2][1].plot(x,sin_values,color = "red" ,label = "sin")
ax[2][1].plot(x,cos_values,color = "green" ,label = "cos")

ax[2][1].fill_between(x,sin_values,cos_values,
                       where = sin_values>=cos_values,color = "chocolate")

ax[2][1].set(title = "using where option",
       xlabel = "x",
       ylabel = "sine values")
ax[2][1].legend(loc = "upper left")


# using interpoldiff = C-S
ax[3][0].plot(x,sin_values,color = "red" ,label = "sin")
ax[3][0].plot(x,cos_values,color = "green" ,label = "cos")

ax[3][0].fill_between(x,sin_values,cos_values,
                       where = sin_values>=cos_values,
                      color = "chocolate",
                      alpha = 0.8,
                      interpolate = True)

ax[3][0].set(title = "using where and interpolate",
       xlabel = "x",
       ylabel = "sine values")
ax[3][0].legend(loc = "upper left")


plt.grid()
plt.show()
```
